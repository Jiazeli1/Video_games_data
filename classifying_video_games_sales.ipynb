```python
feature_sets = [ (Name_features(Name), Publisher) for Name, Publisher in features_and_labels]
feature_sets
```

```python
training_set = feature_sets[:13278]
testing_set = feature_sets[13278:]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
classifier.classify(Name_features(('Sonic Mega Collection Plus')))
```
